### This Melbourne restaurant analysis Project consists in scraping a website Page to parse the necessary information of the names of suburbs in Melbourne, append with longitude/latitude data provided by Australia Government. Then use Foursquare API to get the nearby venues of each of the suburbs. Then using unsupervised machine learning K-means algorithm, cluster the suburbs to analyze each cluster in order to identify the cluster that we need. Then use buildings and restaurants data to identifying the best suburbs from this cluster which are to start restaurant businesses.
